```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as pl
```

## PG - The Procter and Gamble Company

## MSFT - Microsoft Corporation
 
## F - Ford Motor Company
 
## GE - General Electric company

```python
tickets = ['PG' , 'MSFT' , 'F' , 'GE']
mydata = pd.DataFrame()
for t in tickets:
    mydata[t] = wb.DataReader(t, data_source= 'yahoo' , start = '1995-1-1')['Adj Close']
```

```python
mydata
```

```python
mydata.plot(figsize=(10,7))
```

# Calculating The Return of Portfolio Securities

### 1. Calculating the simple rate of return

```python
myreturn= (mydata/mydata.shift(1))-1
```

```python
myreturn
```

### 2. We have the weights of each stock in Portfolio

```python
weights = np.array([0.25,0.25,0.25,0.25])
```

### 3. Doing dot product of returns and weights ------- wait a second return should be average

```python
annual_returns = myreturn.mean() * 250
```

```python
annual_returns
```

```python
np.dot(annual_returns,weights)
```

```python
portfolio1=str(round(np.dot(annual_returns,weights),4)*100) + "%"
```
