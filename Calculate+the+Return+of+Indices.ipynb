```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as pl
```

# MARKET INDEX DATA

### S&P 500   -     ^GSPC

### NASDAQ  -    ^IXIC

### German DAX -  ^GDAXI

### London FTSE - ^FTSE

```python
tickers = ['^GSPC' , '^IXIC' , '^GDAXI' , '^FTSE']
```

```python
ind_data = pd.DataFrame()
for t in tickers:
    ind_data[t] = wb.DataReader(t, data_source='yahoo' , start= '1995-1-1')['Adj Close']
```

```python
ind_data
```

```python
((ind_data/ind_data.iloc[0])*100).plot(figsize=(15,6))
```

```python
pl.show()
```

## Lets see Annual Index Return

```python
index_return = (ind_data/ind_data.shift(1)) - 1
```

```python
annual_index_return = index_return.mean() * 250
```
